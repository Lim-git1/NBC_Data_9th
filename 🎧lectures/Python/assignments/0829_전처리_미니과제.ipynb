## 문제 1 : 단일 열 그룹화 - type1별 평균 공격력
type1 기준으로 그룹화하여 attack의 평균을 구하세요. (결과는 Series/한 컬럼 집계)

```python
pokemon
```

```python
pokemon.columns
```

```python
pokemon.groupby('type1')['attack'].mean()
```

## 문제 2 : 다중 열 그룹화 - generation × is_legendary 조합별 개수
generation과 is_legendary 두 컬럼으로 그룹화하고, 각 그룹의 포켓몬 개수를 구한 뒤 count라는 이름의 컬럼으로 표 형태로 출력하세요.

```python
# genration과 is_legendary -> count => 컬럼명count
result = pokemon.groupby(['generation', 'is_legendary'])['name'].count()
result.reset_index(name='count')
```

## 문제 3번 : 여러 집계 한 번에 - type1별 hp 평균 + attack 평균
type1별로 hp의 평균과 attack의 평균을 동시에 계산하세요. (컬럼별 다른 함수 지정: agg 딕셔너리)

```python
# type1 -> hp , attack .mean
pokemon.groupby('type1').agg({
    'hp':'mean',
    'attack':'mean'
})
```

## 문제 4번 : 문자열 기본: 대문자 변환 + 길이 구하기
name 컬럼을 모두 대문자로 바꾼 name_upper와, 이름 글자 수를 담은 name_len 컬럼을 추가하고, 상위 5개 행만 확인하세요.

```python
name_upper = pokemon['name'].str.upper()
name_len = pokemon['name'].str.len()

pokemon['name_upper'] = name_upper
pokemon['name_len'] = name_len

pokemon.head()
```

## 문제 5번 : 날짜 기본: 문자열 → 날짜 변환 & 연도/요일 추출
아래 날짜 문자열 리스트를 Series로 만들고 pd.to_datetime으로 변환한 뒤, 연도(year)와 요일(day_name)을 각각 출력하세요.

```python
dates_str = ['1996-02-27', '1999-11-21', '2002-11-21']
dates_str
```

```python
# 1. 날짜 문자열 리스트를 Series로 만들기
dates_str = pd.Series(dates_str)
dates_str

# 2. pd.to_datetime으로 변환
dates_datetime = pd.to_datetime(dates_str,format='%Y-%m-%d')
dates_datetime

# 3. 연도(year)와 요일(day_name)을 각각 출력
dates_datetime.dt.year.tolist()
dates_datetime.dt.day_name().tolist()
```

## 문제 6번 : map으로 쉬운 값 치환 - 전설 여부 라벨 만들기
is_legendary가 1이면 "Legendary", 0이면 "Normal"을 부여한 새 컬럼 legendary_label을 만드세요.

```python
pokemon['legebdary_label']= pokemon['is_legendary'].map({1:'Legendary', 0:'Nomal'})
pokemon
```

## 문제 7번 : concat 기초 - Fire와 Water 묶어 세로로 붙이기
type1 == 'Fire'인 포켓몬의 name, type1만 담은 fire_df와 </br>
type1 == 'Water'인 포켓몬의 name, type1만 담은 water_df를 만든 뒤,</br> 두 DataFrame을 세로 방향으로 이어붙여(ignore_index=True) fw_concat을 만들고 상위 8행을 확인하세요.

```python
# DataFrame.loc[행 조건(또는 인덱스), 열 선택]


# 1. (type1 == 'Fire') -> name, type1 => fire_df
fire_df = pokemon.loc[pokemon['type1'] == 'fire', ['name', 'type1']]

# 2. (type1 == 'Water') -> name, type1 => Water_df
water_df = pokemon.loc[pokemon['type1'] == 'water', ['name', 'type1']]

# 3. 두 DataFrame을 세로 방향으로 이어붙여(ignore_index=True) => fw_concat
fw_concat = pd.concat([fire_df, water_df], axis=0, ignore_index=True)

# 4. 상위 8행 출력
fw_concat.head(8)
```

## 문제 8번 : generation 라벨 붙이기 (Left Merge)
아래의 작은 매핑 테이블 gen_info를 직접 만들고, pokemon과 generation 컬럼을 기준으로 left merge 하여 gen_name 컬럼을 붙이세요. 결과의 상위 5행을 출력하세요.

매핑은 다음과 같이 사용하세요:  
1 → "Gen I"  
2 → "Gen II"  
3 → "Gen III"  
4 → "Gen IV"  
5 → "Gen V"  
6 → "Gen VI"  
7 → "Gen VII"  

```python
# 1. 매핑 테이블 gen_info 생성
gen_info = pd.DataFrame({
    'generation' : [1,2,3,4,5,6,7],
    'gen' : ["Gen I", "Gen II" ,"Gen III" ,"Gen IV", "Gen V", "Gen VI", "Gen VII"]
})
gen_info

# 2. pokemon과 generation 컬럼을 기준으로 left merge => gen_name 컬럼을 붙이세요. 
gen_name = pokemon.merge(gen_info, on='generation', how='left')

# 3. 상위 5행을 출력
gen_name.head()
```
