### **선형 그래프**

**언제 사용하지?**  :   시계열 자료, 순서가 있는 데이터 추이

- Matplotlib: `plt.plot(x, y)`

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

days = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
sales = [10, 12, 9, 13, 15, 14, 11]

plt.plot(days, sales, marker="o")
plt.title("Daily Sales")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.show()
```

- Seaborn: `sns.lineplot(x=..., y=..., data=...)`

```python
df = pd.DataFrame({"day": days, "sales": sales})
sns.lineplot(x="day", y="sales", data=df, marker="o")
plt.title("Daily Sales (Seaborn)")
plt.show()
```

### **막대 그래프**

**언제 사용하지?** : 범주/분류에 따라 값을 비교

- Matplotlib : 
  
  - `plt.barh(fruits, stocks)` -> 수평
  
  - `plt.bar(fruits, stocks)` -> 수직

```python
fruits = ["Apple","Banana","Cherry"]
stocks = [30, 18, 25]

# 수평
plt.barh(fruits, stocks) 
# plt.bar(fruits, stocks) -> 수직
plt.title("Fruit Stocks")
plt.xlabel("Count")
plt.show()
```

- Seaborn : `sns.barplot(x=..., y=..., data=...)`

```python
basket = pd.DataFrame({"fruit": ["Apple","Banana","Apple","Cherry","Apple"], "stocks":[30,18,25,27,55]})
sns.barplot(x='fruit', y='stocks', data=basket)
plt.title("Fruit Count (Auto)")
plt.show()
```

### **산점도(Scatter Plot)**

**언제 사용하지?** : 두 변수 사이의 상관관계

- Matplotlib : `plt.scatter(x, y)`

```python
height = [160, 165, 170, 175, 180]
weight = [55, 60, 65, 72, 80]

plt.scatter(height, weight)
plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
```

- Seaborn : `sns.scatterplot(x='col1', y='col2', data=df)`

```python
df = pd.DataFrame({
    "height": [160,165,170,175,180,168,172],
    "weight": [55,60,65,72,80,58,67],
    "sex":    ["F","M","M","M","M","F","F"]
})
sns.scatterplot(x="height", y="weight", hue="sex", data=df)
plt.title("Height vs Weight by Sex")
plt.show()
```

### **히스토그램**

**언제 사용할까?** : 연속형데이터 분포 (어디에 몰려있지? 어떤 구간이 많지?)

Matplotlib: `plt.hist(data, bins=10)`

```python
scores = [55, 60, 62, 65, 70, 72, 75, 78, 80, 82, 85, 88]
plt.hist(scores, bins=5)  # 구간 5개로 쪼개기
plt.title("Score Distribution")
plt.xlabel("Score"); plt.ylabel("Count")
plt.show()
```

Seaborn: `sns.histplot(data=df['some_column'])`

```python
df = pd.DataFrame({"score": scores})
sns.histplot(data=df, x="score", bins=5)  # kde=True(부드러운 곡선) 옵션 가능
plt.title("Score Distribution (Seaborn)")
plt.show()
```

### **박스플롯(Box Plot) & 바이올린 플롯(Violin Plot)**

**언제 사용하지?** : 범주별 데이터의 중앙값, 사분위수, 이상치

* Seaborn 구현이 더 간단하다.

```python
df = pd.DataFrame({
    "class": ["A","A","A","B","B","B"],
    "score": [70, 75, 80, 65, 68, 78]
})
sns.boxplot(x="class", y="score", data=df)
plt.title("Scores by Class (Boxplot)")
plt.show()
```

```python
# order : 순서, inner : 중앙값/사분위선 표시
sns.violinplot(x="class", y="score", data=df, inner="quartile")
plt.title("Scores by Class (Violin)")
plt.show()
```

# IMDB 5000 Movie Dataset 시각화 실습

```python
# plotly 라이브러리 설치
# pip install plotly
```

```python
import pandas as pd
df = pd.read_csv('../data/movie_metadata.csv')
```

## 막대그래프: 연도별 영화 개수

### Matplotlib을 사용한 막대그래프

```python
import matplotlib.pyplot as plt

# 연도별 영화 개수 계산
movies_per_year = df['title_year'].value_counts().sort_index()  # 연도별 count (연도 정렬)
years = movies_per_year.index.astype(int)
counts = movies_per_year.values

# 막대그래프 그리기
plt.figure(figsize=(6,4))
plt.bar(years, counts, color='skyblue', width=0.6)
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies Released Each Year')
plt.xticks(years, rotation=45)
plt.tight_layout()
plt.show()
```

### Seaborn을 사용한 막대그래프

```python
import seaborn as sns

plt.figure(figsize=(6,4))
sns.barplot(x=movies_per_year.index, y=movies_per_year.values, palette='pastel')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies Released Each Year (Seaborn)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Plotly를 사용한 막대그래프

```python
import plotly.express as px

fig = px.bar(
    x=movies_per_year.index,
    y=movies_per_year.values,
    labels={'x':'Year', 'y':'Number of Movies'},
    title='Number of Movies Released Each Year'
)
fig.show()
```

## 산점도: 예산 vs 수익

### Matplotlib을 사용한 산점도

```python
plt.figure(figsize=(6,4))
plt.scatter(df['budget'], df['gross'], color='teal', alpha=0.7) #alpha는 투명도
plt.xlabel('Budget (USD)')
plt.ylabel('Gross Revenue (USD)')
plt.title('Budget vs Gross Revenue')
plt.tight_layout()
plt.show()
```

### Seaborn을 사용한 산점도

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='budget', y='gross', color='purple')
plt.xlabel('Budget (USD)')
plt.ylabel('Gross Revenue (USD)')
plt.title('Budget vs Gross Revenue (Seaborn)')
plt.tight_layout()
plt.show()
```

## 히트맵: 상관관계 행렬

**상관계수(correlation coefficient)란?**

: 두 변수(예: height, weight)가 얼마나 선형적으로 관련이 있는지를 -1 ~ 1 사이의 값으로 표현한 것.

- +1 → 완벽한 양의 상관관계 (한 변수가 커지면 다른 변수도 커짐)

- -1 → 완벽한 음의 상관관계 (한 변수가 커지면 다른 변수는 작아짐)

- 0 → 선형적 관계 없음

### Matplotlib을 사용한 히트맵

```python
# 선택한 수치형 변수들의 상관계수 행렬 계산
cols = ['duration', 'num_voted_users', 'budget', 'gross', 'imdb_score']
corr_matrix = df[cols].corr().values  # 또는 df[cols].corr().values

plt.figure(figsize=(5,4))
im = plt.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Movie Attributes')
# 축 눈금 및 라벨 설정
plt.xticks(range(len(cols)), cols, rotation=45, ha='right')
plt.yticks(range(len(cols)), cols)
# 컬러바 추가
plt.colorbar(im)
plt.tight_layout()
plt.show()
```

### Seaborn을 사용한 히트맵

```python
corr_df = df[cols].corr()  # 상관계수를 DataFrame으로 얻기
plt.figure(figsize=(5,4))
sns.heatmap(corr_df, annot=True, cmap='coolwarm', center=0) # annot : 상관계수표시
plt.title('Correlation Heatmap of Movie Attributes (Seaborn)')
plt.tight_layout()
plt.show()
```
